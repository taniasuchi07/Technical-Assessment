```python
"""
make_figures_from_matrix.py

1) Parses a DOCX containing a literature matrix (appendix) and extracts rows.
2) Infers task domains and evaluation metric mentions from each entry using keyword heuristics.
3) Builds data structures for three data-driven figures:
     - Figure 2: Benchmark Landscape Map  (task -> methods/datasets)
     - Figure 3: Shift in Evaluation Metrics Over Time (metric counts by year)
     - Figure 4: Benchmark Coverage by Domain (domain counts)
4) Draws six figures (1..6) as PNG files.

IMPORTANT:
 - This script computes counts from the docx you provided (no invented numbers).
 - Review console printouts (landscape_map, metric-year counts, domain_counts) to ensure mappings are correct.
 - If mappings need fine-tuning, edit the keyword lists in classify_domain() and detect_metrics().

Usage:
    pip install python-docx matplotlib numpy
    python make_figures_from_matrix.py /path/to/QMH02.docx

Default path used in script: ./QMH02.docx (adjust if necessary)
"""

import sys
import re
from collections import defaultdict, Counter
from docx import Document
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, FancyBboxPatch, ConnectionPatch, ArrowStyle
import numpy as np

# ---------------------------
# 1) Utilities: DOCX table parsing
# ---------------------------
def find_literature_table(doc):
    """
    Find the table which looks like the literature matrix (based on presence of 'Author' and 'Title' in header).
    Returns the docx.table.Table object or raises ValueError.
    """
    for tbl in doc.tables:
        # fetch header row text (first row)
        try:
            header_texts = [cell.text.strip().lower() for cell in tbl.rows[0].cells]
        except Exception:
            continue
        if any('author' in h for h in header_texts) and any('title' in h for h in header_texts):
            return tbl
    raise ValueError("Could not find a table with 'Author' and 'Title' headers. Check the docx layout.")

def table_to_dicts(table):
    """
    Convert a docx table to a list of dicts using header names (normalized).
    Returns list[dict].
    """
    # get header cells
    header_cells = [c.text.strip() for c in table.rows[0].cells]
    # normalize header -> short keys
    def map_header(h):
        h_low = h.lower()
        if 'author' in h_low: return 'author_year'
        if 'title' in h_low: return 'title'
        if 'journal' in h_low or 'journal name' in h_low: return 'journal'
        if 'url' in h_low: return 'url'
        if 'research objectives' in h_low: return 'research_objectives'
        if 'research questions' in h_low: return 'research_questions'
        if 'problems' in h_low or 'issues' in h_low: return 'problems'
        if 'theories' in h_low or 'models' in h_low: return 'theories_models'
        if 'variables' in h_low: return 'variables'
        if 'findings' in h_low: return 'findings'
        if 'conclusion' in h_low: return 'conclusion'
        if 'limitations' in h_low: return 'limitations'
        if 'respondents' in h_low: return 'respondents'
        if 'statistical' in h_low: return 'statistical_technique'
        # fallback: slugify header
        key = re.sub(r'[^\w]+', '_', h_low).strip('_')
        return key or f'col_{len(header_cells)}'

    keys = [map_header(h) for h in header_cells]

    rows = []
    for row in table.rows[1:]:
        cells = [c.text.strip() for c in row.cells]
        # if number of cells mismatches, pad or trim
        if len(cells) < len(keys):
            cells += [''] * (len(keys) - len(cells))
        cells = cells[:len(keys)]
        d = {k: v for k, v in zip(keys, cells)}
        # attempt to extract year
        year = None
        if 'author_year' in d:
            m = re.search(r'\b(19|20)\d{2}\b', d['author_year'])
            if m:
                year = int(m.group(0))
        if not year:
            # try title
            m = re.search(r'\b(19|20)\d{2}\b', d.get('title',''))
            if m:
                year = int(m.group(0))
        d['year'] = year
        rows.append(d)
    return rows

# ---------------------------
# 2) Heuristics: classify domain and detect metrics from text
# ---------------------------
def classify_domain(row):
    """
    Heuristic rules to classify each row into a domain category:
      - Summarization, Dialogue, QA / Factuality, RAG, General Surveys, Other
    """
    text = " ".join([row.get('title',''), row.get('research_objectives',''), row.get('problems',''), row.get('findings','')]).lower()
    if 'summar' in text or 'summarization' in text or 'summaries' in text:
        return 'Summarization'
    if 'dialog' in text or 'multi-turn' in text or 'dialogue' in text:
        return 'Dialogue'
    if 'qa' in text or 'question' in text or 'truthful' in text or 'entity' in text or 'factual' in text or 'fact' in text:
        return 'QA / Factuality'
    if 'rag' in text or 'retriev' in text or 'retrieval' in text or 'attribut' in text:
        return 'RAG'
    if 'survey' in text or 'review' in text or 'systematic review' in text or 'surveyed' in text:
        return 'General Surveys'
    # fallback to other
    return 'Other'

def detect_metrics(row):
    """
    Detect evaluation/metric types mentioned in the row text.
    Returns a set of metric labels from:
      ROUGE/BLEU, BERTScore, Entailment/QA, LLM-as-Judge, Metamorphic, Survey Reviews, BERT embeddings (same as BERTScore)
    """
    text = " ".join([row.get('title',''), row.get('research_objectives',''), row.get('problems',''), row.get('findings',''), row.get('theories_models','')]).lower()
    metrics = set()
    if any(k in text for k in ['rouge', 'bleu', 'meteor']):
        metrics.add('ROUGE/BLEU')
    if 'bertscore' in text or 'bert score' in text or 'embedding' in text:
        metrics.add('BERTScore')
    if any(k in text for k in ['entail', 'qa-based', 'qa ', 'exact match', 'f1', 'question answering', 'truthfulqa', 'truthful']):
        metrics.add('Entailment/QA')
    if 'llm-as-judge' in text or ('judge' in text and 'llm' in text) or ('llm as judge' in text):
        metrics.add('LLM-as-Judge')
    if any(k in text for k in ['metamorphic', 'mutation', 'prompt mutation', 'metamorph']):
        metrics.add('Metamorphic')
    if 'survey' in text or 'review' in text or 'surveyed' in text:
        metrics.add('Survey Reviews')
    # special: 'BERTScore Team' or 'BERTScore' is BERTScore
    return metrics

# ---------------------------
# 3) Build landscape_map, metric time series, domain counts
# ---------------------------
def build_structures(rows, min_year=2020, max_year=2025):
    """
    From parsed rows, build:
      - landscape_map: dict task -> set(method strings found)
      - metric_year_counts: dict metric -> Counter(year->count)
      - domain_counts: Counter(domain->count)
    """
    landscape_map = defaultdict(set)
    metric_year_counts = defaultdict(Counter)
    domain_counter = Counter()
    seen_entries_for_landscape = defaultdict(list)  # track which papers added which method

    for r in rows:
        domain = classify_domain(r)
        domain_counter[domain] += 1

        # add example method descriptions (short) to landscape_map using keys from row
        # We'll add entries based on titles / findings / theoretical hints
        title = r.get('title','')
        # create simple short label
        short_label = title if title else r.get('author_year','').split('\n')[0]

        # heuristics to add specific method labels
        methods = []
        t = " ".join([r.get('title',''), r.get('research_objectives',''), r.get('problems',''), r.get('findings',''), r.get('theories_models','')]).lower()
        if 'summar' in t:
            methods.append('Summarization benchmarks (e.g., SummEval)')
        if 'dialog' in t:
            methods.append('Dialogue-level benchmark (e.g., DiaHalu)')
        if 'truthful' in t or 'truth' in t or 'question' in t or 'qa' in t:
            methods.append('QA-based probes (TruthfulQA, WildHallucinations)')
        if 'bert' in t or 'bertscore' in t:
            methods.append('Embedding-based metrics (BERTScore)')
        if 'entail' in t:
            methods.append('Entailment / NLI probes')
        if 'metamorphic' in t or 'mutation' in t or 'metaqa' in t:
            methods.append('Metamorphic / prompt-mutation testing')
        if 'rag' in t or 'retriev' in t or 'attribut' in t:
            methods.append('RAG grounding & retrieval checks')
        if 'survey' in t or 'review' in t:
            methods.append('Survey / metric review')

        # fallback: if no method detected, add short label under its domain
        if not methods:
            methods.append(short_label)

        for m in methods:
            landscape_map[domain].add(m)
            seen_entries_for_landscape[domain].append((short_label, m))

        # metrics per row -> increment year counts
        metrics = detect_metrics(r)
        year = r.get('year')
        if year is None:
            # skip year-based metrics if unknown
            continue
        # limit to user-specified year range (default 2020-2025 inclusive)
        if year < min_year or year > max_year:
            # still allow counting, but they won't appear in the default plotting range unless you change min_year/max_year
            pass
        for metric in metrics:
            metric_year_counts[metric][year] += 1

    return landscape_map, metric_year_counts, domain_counter, seen_entries_for_landscape

# ---------------------------
# # 4) Plotting functions (Figures 1-6)
# ---------------------------
def plot_conceptual_triad(savepath="figure1_triad.png", title="Conceptual Triad: Factuality, Faithfulness, Hallucination"):
    fig, ax = plt.subplots(figsize=(8, 6))
    ax.set_title(title, fontsize=14, pad=14)
    colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
    r = 1.3
    centers = [(-0.8, 0), (0.8, 0), (0, 1.0)]
    c1 = Circle(centers[0], r, alpha=0.25, facecolor=colors[0], edgecolor='k')
    c2 = Circle(centers[1], r, alpha=0.25, facecolor=colors[1], edgecolor='k')
    c3 = Circle(centers[2], r, alpha=0.25, facecolor=colors[2], edgecolor='k')
    for c in (c1,c2,c3):
        ax.add_patch(c)
    ax.text(centers[0][0], centers[0][1], "Factuality\n(external truth)", ha='center', va='center', fontsize=10)
    ax.text(centers[1][0], centers[1][1], "Faithfulness\n(to source/input)", ha='center', va='center', fontsize=10)
    ax.text(centers[2][0], centers[2][1], "Hallucination\n(unsupported claims)", ha='center', va='center', fontsize=10)
    ax.text(0, 0.15, "Factual & Faithful\n(ideal)", ha='center', va='center', fontsize=9)
    ax.text(-0.35, -0.55, "Factual but\nNot Faithful", ha='center', va='center', fontsize=8)
    ax.text(0.45, -0.55, "Faithful but\nNot Factual", ha='center', va='center', fontsize=8)
    ax.text(0.0, 1.25, "Hallucinated\nClaims", ha='center', va='center', fontsize=8)
    ax.axis('equal')
    ax.axis('off')
    fig.savefig(savepath, dpi=300, bbox_inches='tight')
    print(f"Saved Figure 1 -> {savepath}")
    return fig

def plot_benchmark_landscape(landscape_map, savepath="figure2_landscape.png", title="Benchmark Landscape Map"):
    # convert sets to lists and sort for deterministic output
    items = list(landscape_map.items())
    fig, ax = plt.subplots(figsize=(10, max(4, 1 + 0.8 * len(items))))
    ax.set_title(title, fontsize=14, pad=12)
    left_x = 0.04
    right_x = 0.48
    y_start = 0.9
    y_step = 1.0 / (len(items) + 0.5)
    box_height = 0.06
    for i, (task, methods) in enumerate(items):
        y = y_start - i * y_step
        task_box = FancyBboxPatch((left_x, y - box_height/2), 0.38, box_height, boxstyle="round,pad=0.02", linewidth=1)
        ax.add_patch(task_box)
        ax.text(left_x + 0.02, y, task, va='center', ha='left', fontsize=10, weight='bold')
        methods_text = "\n".join(f"- {m}" for m in sorted(methods))
        methods_box = FancyBboxPatch((right_x, y - box_height/2 - 0.03), 0.48, box_height + 0.06, boxstyle="round,pad=0.02", linewidth=1)
        ax.add_patch(methods_box)
        ax.text(right_x + 0.02, y + 0.03, methods_text, va='top', ha='left', fontsize=9)
        ax.annotate("", xy=(left_x + 0.38, y), xytext=(right_x, y), arrowprops=dict(arrowstyle="-", linewidth=1))
    ax.set_xlim(0,1)
    ax.set_ylim(0,1)
    ax.axis('off')
    fig.savefig(savepath, dpi=300, bbox_inches='tight')
    print(f"Saved Figure 2 -> {savepath}")
    return fig

def plot_metric_shift_over_time(metric_year_counts, savepath="figure3_metric_shift.png", title="Shift in Evaluation Metrics Over Time", years=None):
    # Determine the years range from the metric_year_counts if not provided
    all_years = set()
    for cnt in metric_year_counts.values():
        all_years.update(cnt.keys())
    if years is None:
        if all_years:
            years = sorted(all_years)
        else:
            # fallback to 2020-2025
            years = list(range(2020, 2026))
    fig, ax = plt.subplots(figsize=(9,5))
    ax.set_title(title, fontsize=14)
    for metric, cnt in metric_year_counts.items():
        values = [cnt.get(y, 0) for y in years]
        ax.plot(years, values, marker='o', label=metric)
    ax.set_xlabel("Year")
    ax.set_ylabel("Count (papers mentioning metric)")
    ax.legend()
    ax.grid(axis='y', linestyle='--', linewidth=0.4)
    fig.savefig(savepath, dpi=300, bbox_inches='tight')
    print(f"Saved Figure 3 -> {savepath}")
    return fig

def plot_coverage_by_domain(domain_counts, savepath="figure4_domain_coverage.png", title="Benchmark Coverage by Domain"):
    domains = list(domain_counts.keys())
    values = [domain_counts[d] for d in domains]
    fig, ax = plt.subplots(figsize=(8,5))
    ax.set_title(title, fontsize=14)
    bars = ax.bar(domains, values)
    ax.set_ylabel("Number of papers")
    ax.set_xticklabels(domains, rotation=25, ha='right')
    ax.grid(axis='y', linestyle='--', linewidth=0.4)
    for bar, val in zip(bars, values):
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.3, str(val), ha='center', va='bottom', fontsize=9)
    fig.savefig(savepath, dpi=300, bbox_inches='tight')
    print(f"Saved Figure 4 -> {savepath}")
    return fig

def plot_hallucination_workflow(savepath="figure5_workflow.png", title="Hallucination Detection Workflow"):
    fig, ax = plt.subplots(figsize=(10, 3.5))
    ax.set_title(title, fontsize=14, pad=10)
    boxes = [
        ("Input\n(document(s) / prompt)", (0.05, 0.5)),
        ("LLM\n(generation)", (0.30, 0.5)),
        ("Automatic probes\n(QA / entailment)", (0.55, 0.5)),
        ("LLM-as-judge\n(or human judge)", (0.80, 0.5)),
    ]
    box_w = 0.22
    box_h = 0.25
    for label, (x, y) in boxes:
        rect = FancyBboxPatch((x, y - box_h/2), box_w, box_h, boxstyle="round,pad=0.03", linewidth=1)
        ax.add_patch(rect)
        ax.text(x + box_w/2, y, label, ha='center', va='center', fontsize=9)
    for i in range(len(boxes)-1):
        start = (boxes[i][1][0] + box_w, boxes[i][1][1])
        end = (boxes[i+1][1][0], boxes[i+1][1][1])
        con = ConnectionPatch(start, end, "data", "data", arrowstyle=ArrowStyle("-|>", head_length=6, head_width=4), linewidth=1)
        ax.add_artist(con)
    ax.annotate("Flagged", xy=(0.62, 0.35), xytext=(0.62, 0.15),
                arrowprops=dict(arrowstyle="->", linewidth=1))
    ax.text(0.62, 0.12, "Human review / evidence-check", ha='center', va='center', fontsize=8)
    ax.axis('off')
    fig.savefig(savepath, dpi=300, bbox_inches='tight')
    print(f"Saved Figure 5 -> {savepath}")
    return fig

def plot_rag_reliability_framework(savepath="figure6_rag_framework.png", title="RAG Reliability Framework"):
    fig, ax = plt.subplots(figsize=(10, 3.5))
    ax.set_title(title, fontsize=14, pad=10)
    stages = [
        ("Retriever\n(recall/precision)", 0.02),
        ("Retriever\n(freshness & bias)", 0.26),
        ("Context\n(quality & provenance)", 0.50),
        ("Generator\n(grounding checks)", 0.74),
        ("Attribution\n(citations/links)", 0.90),
    ]
    box_w = 0.22
    box_h = 0.24
    for label, x in stages:
        rect = FancyBboxPatch((x, 0.55 - box_h/2), box_w, box_h, boxstyle="round,pad=0.03", linewidth=1)
        ax.add_patch(rect)
        ax.text(x + box_w/2, 0.55, label, ha='center', va='center', fontsize=9)
    for i in range(len(stages)-1):
        start = (stages[i][1] + box_w, 0.55)
        end = (stages[i+1][1], 0.55)
        con = ConnectionPatch(start, end, "data", "data", arrowstyle=ArrowStyle("-|>", head_length=6, head_width=4), linewidth=1)
        ax.add_artist(con)
    eval_box = FancyBboxPatch((0.40, 0.08), 0.22, 0.18, boxstyle="round,pad=0.03", linewidth=1)
    ax.add_patch(eval_box)
    ax.text(0.51, 0.17, "Evaluation\n(coverage, faithfulness,\nsource verifiability)", ha='center', va='center', fontsize=9)
    con2 = ConnectionPatch((0.61, 0.55), (0.61, 0.26), "data", "data", arrowstyle="->", linewidth=1)
    ax.add_artist(con2)
    ax.axis('off')
    fig.savefig(savepath, dpi=300, bbox_inches='tight')
    print(f"Saved Figure 6 -> {savepath}")
    return fig

# ---------------------------
# 5) Main: parse file, build structures, plot figures
# ---------------------------
def main(docx_path="QMH02.docx"):
    print(f"Reading DOCX: {docx_path}")
    doc = Document(docx_path)
    table = find_literature_table(doc)
    rows = table_to_dicts(table)
    print(f"Parsed {len(rows)} rows from the literature matrix table.")

    # Build data structures (default year window 2020-2025)
    landscape_map, metric_year_counts, domain_counter, seen_entries = build_structures(rows, min_year=2020, max_year=2025)

    # print summaries so you can verify
    print("\n--- Landscape map overview (domain -> methods) ---")
    for domain, methods in landscape_map.items():
        print(f"{domain}:")
        for m in sorted(methods):
            print(f"  - {m}")

    print("\n--- Domain counts ---")
    for d, c in domain_counter.items():
        print(f" {d}: {c}")

    print("\n--- Metric counts by year ---")
    # show metric -> year -> count
    for metric, cnt in metric_year_counts.items():
        print(f"{metric}: {dict(sorted(cnt.items()))}")

    # Now plot the six figures:
    plot_conceptual_triad(savepath="figure1_triad.png")
    plot_benchmark_landscape(landscape_map, savepath="figure2_landscape.png")
    plot_metric_shift_over_time(metric_year_counts, savepath="figure3_metric_shift.png")
    plot_coverage_by_domain(domain_counter, savepath="figure4_domain_coverage.png")
    plot_hallucination_workflow(savepath="figure5_workflow.png")
    plot_rag_reliability_framework(savepath="figure6_rag_framework.png")
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon, Circle, FancyArrowPatch, Rectangle
import numpy as np

def draw_triadic_reliability_matrix(save_png=True, save_svg=False, fname_base="triadic_reliability_matrix"):
    # --- Geometry: equilateral triangle ---
    # Vertices: Hallucination (top), Factuality (left base), Faithfulness (right base)
    H = np.array([0.5, 0.92])
    FCT = np.array([0.08, 0.08])
    FTH = np.array([0.92, 0.08])

    fig, ax = plt.subplots(figsize=(8, 8))
    ax.set_aspect('equal')
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.axis('off')

    # Triangle outline
    tri = Polygon([FCT, FTH, H], closed=True, fill=False, linewidth=2.0, edgecolor='black')
    ax.add_patch(tri)

    # Soft background tint
    tri_bg = Polygon([FCT, FTH, H], closed=True, fill=True, linewidth=0, facecolor="#f7f7f7", alpha=1.0)
    ax.add_patch(tri_bg)

    # --- Regions (semi-qualitative zones) ---
    # Reliable zone: near base center (high Factuality & Faithfulness, low Hallucination)
    reliable_center = (FCT + FTH) / 2
    reliable = Circle(reliable_center + np.array([0.0, 0.03]), radius=0.08,
                      facecolor="#a1d99b", edgecolor="black", linewidth=1.0, alpha=0.7)
    ax.add_patch(reliable)
    ax.text(reliable_center[0], reliable_center[1] + 0.03,
            "Reliable Region\n(Factual + Faithful,\nLow Hallucination)",
            ha="center", va="center", fontsize=11, weight="bold")

    # Factual but Unfaithful: left base area
    left_zone = Polygon([
        FCT,
        (FCT + FTH) / 2,
        (FCT + H) / 2
    ], closed=True, facecolor="#c7e9c0", edgecolor="black", linewidth=0.8, alpha=0.6)
    ax.add_patch(left_zone)
    ax.text(0.22, 0.22, "Factual but\nUnfaithful", ha="center", va="center", fontsize=10)

    # Faithful but Factually Wrong: right base area
    right_zone = Polygon([
        FTH,
        (FCT + FTH) / 2,
        (FTH + H) / 2
    ], closed=True, facecolor="#c6dbef", edgecolor="black", linewidth=0.8, alpha=0.6)
    ax.add_patch(right_zone)
    ax.text(0.78, 0.22, "Faithful but\nFactually Wrong", ha="center", va="center", fontsize=10)

    # Hallucination-heavy zone: towards the top
    top_zone = Polygon([
        (FCT + H) / 2,
        (FTH + H) / 2,
        H
    ], closed=True, facecolor="#fcbba1", edgecolor="black", linewidth=0.8, alpha=0.6)
    ax.add_patch(top_zone)
    ax.text(0.5, 0.72, "Both Unfaithful +\nHallucinated", ha="center", va="center", fontsize=10)

    # Optional sub-zones near top-left and top-right to show flavors of hallucination contexts
    # (not strict sets; purely illustrative overlays)
    tl_zone = Polygon([
        (FCT + H) / 2,
        (FCT*0.6 + H*0.4),
        (FCT*0.4 + H*0.6)
    ], closed=True, facecolor="#fdd0a2", edgecolor="black", linewidth=0.6, alpha=0.5)
    ax.add_patch(tl_zone)
    ax.text(0.30, 0.58, "Factual\nHallucination", ha="center", va="center", fontsize=9)

    tr_zone = Polygon([
        (FTH + H) / 2,
        (FTH*0.6 + H*0.4),
        (FTH*0.4 + H*0.6)
    ], closed=True, facecolor="#f2b6c6", edgecolor="black", linewidth=0.6, alpha=0.5)
    ax.add_patch(tr_zone)
    ax.text(0.70, 0.58, "Faithful\nHallucination", ha="center", va="center", fontsize=9)

    # --- Axes labels at vertices ---
    def place_vertex_label(pt, text, dx=0, dy=0):
        ax.text(pt[0]+dx, pt[1]+dy, text, ha='center', va='center',
                fontsize=13, weight="bold")

    place_vertex_label(H, "Hallucination ", dy=0.045)
    place_vertex_label(FCT, "Factuality ", dy=-0.045)
    place_vertex_label(FTH, "Faithfulness ", dy=-0.045)

    # Base axis guide (low hallucination near base)
    ax.text(0.5, 0.02, "Lower Hallucination", ha="center", va="center", fontsize=9, color="gray")

    # --- Intrinsic vs Extrinsic Hallucination callouts ---
    # Intrinsic: internal contradictions (box + arrow into top zone)
    box1 = Rectangle((0.03, 0.78), 0.34, 0.15, facecolor="white", edgecolor="black", linewidth=1)
    ax.add_patch(box1)
    ax.text(0.20, 0.855,
            "Intrinsic Hallucination:\nInternal inconsistencies\n(e.g., conflicting values)",
            ha="center", va="center", fontsize=9)
    arr1 = FancyArrowPatch(posA=(0.37, 0.80), posB=(0.52, 0.69),
                           arrowstyle='->', mutation_scale=12, color="black", linewidth=1)
    ax.add_patch(arr1)

    # Extrinsic: unsupported by source/knowledge (box + arrow into top zone)
    box2 = Rectangle((0.63, 0.78), 0.34, 0.15, facecolor="white", edgecolor="black", linewidth=1)
    ax.add_patch(box2)
    ax.text(0.80, 0.855,
            "Extrinsic Hallucination:\nUnsupported by input/KB\n(e.g., fabricated facts)",
            ha="center", va="center", fontsize=9)
    arr2 = FancyArrowPatch(posA=(0.63, 0.80), posB=(0.48, 0.69),
                           arrowstyle='->', mutation_scale=12, color="black", linewidth=1)
    ax.add_patch(arr2)

    # --- Caption ---
    ax.text(0.5, -0.04,
            "Figure: Triadic Reliability Matrix (TRM). Reliability sits near the base center where factuality and faithfulness are high\n"
            "and hallucination is low. Overlap scenarios (left/right) illustrate Factual-but-Unfaithful and Faithful-but-Factually-Wrong.\n"
            "Top region captures compounded risk where outputs diverge from inputs and introduce fabrications. Intrinsic vs Extrinsic\n"
            "hallucination are illustrated as callouts.",
            ha="center", va="top", fontsize=9)

    plt.tight_layout()

    if save_png:
        plt.savefig(f"{fname_base}.png", dpi=300, bbox_inches="tight")
    if save_svg:
        plt.savefig(f"{fname_base}.svg", dpi=300, bbox_inches="tight")

    plt.show()

# Run once to generate the figure files
if __name__ == "__main__":
    draw_triadic_reliability_matrix(save_png=True, save_svg=False)
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon, Circle, FancyArrowPatch, Rectangle
import numpy as np

def draw_triadic_reliability_matrix(save_png=True, save_svg=False, fname_base="triadic_reliability_matrix"):
    # --- Geometry: equilateral triangle ---
    H = np.array([0.5, 0.92])  # Hallucination (top)
    FCT = np.array([0.08, 0.08])  # Factuality (left base)
    FTH = np.array([0.92, 0.08])  # Faithfulness (right base)

    fig, ax = plt.subplots(figsize=(10, 9))
    ax.set_aspect('equal')
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.axis('off')

    # Triangle outline
    tri = Polygon([FCT, FTH, H], closed=True, fill=False, linewidth=2.0, edgecolor='black')
    ax.add_patch(tri)

    # Soft background tint
    tri_bg = Polygon([FCT, FTH, H], closed=True, fill=True, linewidth=0, facecolor="#f7f7f7", alpha=1.0)
    ax.add_patch(tri_bg)

    # --- Regions ---
    # Reliable zone: near base center
    reliable_center = (FCT + FTH) / 2
    reliable = Circle(reliable_center + np.array([0.0, 0.03]), radius=0.08,
                      facecolor="#a1d99b", edgecolor="black", linewidth=1.0, alpha=0.7)
    ax.add_patch(reliable)
    ax.text(reliable_center[0], reliable_center[1] + 0.03,
            "Reliable Region\n(Factual + Faithful,\nLow Hallucination)",
            ha="center", va="center", fontsize=10, weight="bold")

    # Factual but Unfaithful: left base area
    left_zone = Polygon([
        FCT,
        (FCT + FTH) / 2,
        (FCT + H) / 2
    ], closed=True, facecolor="#c7e9c0", edgecolor="black", linewidth=0.8, alpha=0.6)
    ax.add_patch(left_zone)
    ax.text(0.22, 0.22, "Factual but\nUnfaithful", ha="center", va="center", fontsize=10)

    # Faithful but Factually Wrong: right base area
    right_zone = Polygon([
        FTH,
        (FCT + FTH) / 2,
        (FTH + H) / 2
    ], closed=True, facecolor="#c6dbef", edgecolor="black", linewidth=0.8, alpha=0.6)
    ax.add_patch(right_zone)
    ax.text(0.78, 0.22, "Faithful but\nFactually Wrong", ha="center", va="center", fontsize=10)

    # Top zone: Both Unfaithful + Hallucinated
    top_zone = Polygon([
        (FCT + H) / 2,
        (FTH + H) / 2,
        H
    ], closed=True, facecolor="#fcbba1", edgecolor="black", linewidth=0.8, alpha=0.6)
    ax.add_patch(top_zone)
    ax.text(0.5, 0.72, "Both Unfaithful +\nHallucinated", ha="center", va="center", fontsize=10)

    # Sub-zones near top-left and top-right
    tl_zone = Polygon([
        (FCT + H) / 2,
        (FCT*0.6 + H*0.4),
        (FCT*0.4 + H*0.6)
    ], closed=True, facecolor="#fdd0a2", edgecolor="black", linewidth=0.6, alpha=0.5)
    ax.add_patch(tl_zone)
    ax.text(0.30, 0.58, "Factual\nHallucination", ha="center", va="center", fontsize=9)

    tr_zone = Polygon([
        (FTH + H) / 2,
        (FTH*0.6 + H*0.4),
        (FTH*0.4 + H*0.6)
    ], closed=True, facecolor="#f2b6c6", edgecolor="black", linewidth=0.6, alpha=0.5)
    ax.add_patch(tr_zone)
    ax.text(0.70, 0.58, "Faithful\nHallucination", ha="center", va="center", fontsize=9)

    # --- Axes labels at vertices ---
    def place_vertex_label(pt, text, dx=0, dy=0):
        ax.text(pt[0]+dx, pt[1]+dy, text, ha='center', va='center',
                fontsize=13, weight="bold")

    place_vertex_label(H, "Hallucination ", dy=0.045)
    place_vertex_label(FCT, "Factuality ", dy=-0.045)
    place_vertex_label(FTH, "Faithfulness ", dy=-0.045)

    # Base axis guide
    ax.text(0.5, 0.02, "Lower Hallucination", ha="center", va="center", fontsize=9, color="gray")

    # --- Intrinsic vs Extrinsic Hallucination callouts ---
    box1 = Rectangle((0.03, 0.78), 0.34, 0.15, facecolor="white", edgecolor="black", linewidth=1)
    ax.add_patch(box1)
    ax.text(0.20, 0.855,
            "Intrinsic Hallucination:\nInternal inconsistencies\n(e.g., conflicting values)",
            ha="center", va="center", fontsize=9)
    arr1 = FancyArrowPatch(posA=(0.37, 0.80), posB=(0.52, 0.69),
                           arrowstyle='->', mutation_scale=12, color="black", linewidth=1)
    ax.add_patch(arr1)

    box2 = Rectangle((0.63, 0.78), 0.34, 0.15, facecolor="white", edgecolor="black", linewidth=1)
    ax.add_patch(box2)
    ax.text(0.80, 0.855,
            "Extrinsic Hallucination:\nUnsupported by input/KB\n(e.g., fabricated facts)",
            ha="center", va="center", fontsize=9)
    arr2 = FancyArrowPatch(posA=(0.63, 0.80), posB=(0.48, 0.69),
                           arrowstyle='->', mutation_scale=12, color="black", linewidth=1)
    ax.add_patch(arr2)

    # --- Caption: Inside the figure, below the triangle ---
    caption_text = (
        "Triadic Reliability Matrix (TRM). Reliability sits near the base center where factuality and faithfulness are high\n"
        "and hallucination is low. Overlap scenarios (left/right) illustrate Factual-but-Unfaithful and Faithful-but-Factually-Wrong.\n"
        "Top region captures compounded risk where outputs diverge from inputs and introduce fabrications. Intrinsic vs Extrinsic\n"
        "hallucination are illustrated as callouts."
    )
    ax.text(0.5, -0.04, caption_text, ha="center", va="top", fontsize=9, color="black", wrap=True)

    plt.tight_layout()

    if save_png:
        plt.savefig(f"{fname_base}.png", dpi=300, bbox_inches="tight")
    if save_svg:
        plt.savefig(f"{fname_base}.svg", dpi=300, bbox_inches="tight")

    plt.show()

# Run once to generate the figure files
if __name__ == "__main__":
    draw_triadic_reliability_matrix(save_png=True, save_svg=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# ------------------------------
# Benchmark Evaluation Framework Scores
# ------------------------------
data = {
    "Benchmark": ["ROUGE/BLEU", "BERTScore", "TruthfulQA", "DiaHalu",
                  "MetaQA", "WildHallucinations", "SummEval", "RAG Monitor"],
    "Scalability": [2, 1, 1, 1, 2, 1, 1, 2],
    "Accuracy":    [0, 2, 2, 1, 1, 2, 1, 1],
    "Cost":        [2, 2, 0, 1, 2, 0, 1, 2],
    "Coverage":    [1, 1, 0, 0, 0, 1, 1, 0]
}
df = pd.DataFrame(data)
df.set_index("Benchmark", inplace=True)

# ------------------------------
# 1. Heatmap: Benchmarks  BEF Dimensions
# ------------------------------
plt.figure(figsize=(8,6))
sns.heatmap(df, annot=True, cmap="RdYlGn", cbar_kws={'label': 'Score (02)'}, linewidths=0.5)
plt.title("Benchmark Evaluation Framework (BEF) Scores", fontsize=14, fontweight="bold")
plt.ylabel("Benchmark")
plt.xlabel("Evaluation Dimension")
plt.tight_layout()
plt.show()

# ------------------------------
# 2. Radar Charts: ROUGE vs BERTScore vs TruthfulQA
# ------------------------------
def make_radar(df, benchmarks, title):
    categories = list(df.columns)
    N = len(categories)

    angles = np.linspace(0, 2*np.pi, N, endpoint=False).tolist()
    angles += angles[:1]  # loop back

    fig, ax = plt.subplots(figsize=(6,6), subplot_kw=dict(polar=True))

    for bench in benchmarks:
        values = df.loc[bench].tolist()
        values += values[:1]
        ax.plot(angles, values, linewidth=2, label=bench)
        ax.fill(angles, values, alpha=0.25)

    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(categories, fontsize=9)
    ax.set_yticks([0,1,2])
    ax.set_yticklabels(["0","1","2"], fontsize=9)
    ax.set_ylim(0,2)
    #ax.set_title(title, fontsize=12, fontweight="bold")
    ax.legend(loc="upper right", bbox_to_anchor=(1.1, 1.1))
    plt.tight_layout()
    plt.show()

make_radar(df, ["ROUGE/BLEU", "BERTScore", "TruthfulQA"],
           "Radar Chart: Trade-offs Across Metrics")

# ------------------------------
# 3. Timeline: Benchmarks Introduced by Year & Domain
# ------------------------------
timeline_data = pd.DataFrame({
    "Year": [2003, 2019, 2021, 2021, 2021, 2022, 2023, 2023, 2024, 2024, 2024],
    "Benchmark": ["ROUGE/BLEU", "BERTScore", "TruthfulQA", "SummEval", "Factual Consistency",
                  "WildHallucinations", "DiaHalu", "MetaQA", "Hallucination Survey",
                  "RAG Monitor", "Faithfulness Metrics Review"],
    "Domain": ["Summarization", "Summarization", "QA", "Summarization", "Summarization",
               "QA", "Dialogue", "QA", "Survey", "RAG", "Survey"]
})

plt.figure(figsize=(8,5))
sns.scatterplot(data=timeline_data, x="Year", y="Domain", hue="Domain", style="Domain", s=120)
plt.title("Timeline of Benchmark (20032024)", fontsize=14, fontweight="bold")
plt.xlabel("Year")
plt.ylabel("Domain")
plt.xticks(range(2003,2025,2))
plt.legend(title="Domain", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
